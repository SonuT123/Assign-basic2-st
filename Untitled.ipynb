```python
Outliers can significantly affect measures of central tendency and dispersion in a dataset. 

1.Measures of Central Tendency:
    
    Mean:-Outliers can have a substantial impact on the mean (average). An outlier with an extremely high or low value can pull the mean 
          in its direction.
         If there is a high positive outlier, the mean will be higher, and if there is a low negative outlier, the mean will be lower.
            
   Median: The median is less affected by outliers than the mean. It represents the middle value when the data is ordered,
            so it is less influenced by extreme values.
           However, a single extreme outlier can still affect the median to some extent, especially in small datasets.  
        
   Mode: The mode is generally less affected by outliers, as it represents the most frequently occurring value(s) in the dataset.
         Outliers usually have a minimal impact on the mode.
    
Example:
Consider a dataset of salaries in a company:

[
40
,
000
,
42
,
000
,
43
,
000
,
45
,
000
,
150
,
000
]
[40,000,42,000,43,000,45,000,150,000]

Mean: The mean is significantly influenced by the high outlier (150,000). 
      The mean salary is skewed upward due to this outlier.

Median: The median is less affected by outliers. In this case,
       the median salary is 43,000, which is closer to the typical salary range of the majority of employees.

Mode: The mode remains the same, which is unaffected by the outlier. 
      The mode is 43,000, representing the most common salary in the dataset. 
    
Measures of Dispersion:

Range: Outliers can have a substantial impact on the range, as the range is determined by the difference between the maximum
       and minimum values in the dataset. The presence of outliers can lead to an overestimation of the range.

Variance and Standard Deviation: Outliers can increase the variance and standard deviation,
      as these measures are based on the squared differences between data points and the mean. 
      Large differences between outliers and the mean contribute significantly to the dispersion.
        
        
Example:
Consider the same dataset of salaries:

[
40
,
000
,
42
,
000
,
43
,
000
,
45
,
000
,
150
,
000
]
[40,000,42,000,43,000,45,000,150,000]

Range: The range is greatly affected by the high outlier, resulting in a larger range.

Variance and Standard Deviation: The variance and standard deviation increase due to the influence of the outlier. 
           These measures reflect the increased variability caused by the outlier.
    
```
